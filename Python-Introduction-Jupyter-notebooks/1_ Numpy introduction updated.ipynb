Numpy allows to write vectors as arrays

To create an interval when knowing its bounds, its easier to use linspace, which precises a number of points instead of a step.
Here we specify 50 points regularly covering the space between 0 and 10.


```python
X = np.linspace(0., 10., 50)
X
```

This would be usefull to graph a function between two bounds. Lets create an array Y containing cosines of X

```python
import matplotlib.pyplot as plt
#%matplotlib inline
plt.ion()
Y = np.cos(X)
plt.plot(X, Y);
```

So numpy makes it very easy to perform elementwise operations on vectors and matrix (nparrays).  
Then if we want to graph the function
$x \longrightarrow cos^2(x)$ and store its results in a array Z:

```python
Z = np.cos(X)**2 
print(Z)
plt.plot(X, Z);
```

Another very usefull attribute of numpy arrays is shape. It gives a dimension (rows, columns).
Let's create a numpy array of dimension 2 from a list

```python
d2 = np.array([[11, 12, 13], [21, 22, 23]])
print(d2)
print(d2.shape)
```

An array is in fact a view of a block of data. So its dimensions are easier to reshape, which is basicallly just creating another view to the same block.

```python
v2 = d2.reshape((3, 2))
v2
```

Be carefull of sides effects.
d2 and v2 are 2 views of the same block, so changing data in one has consequences on the other

```python
d2[0][0] = 100

print(d2)
print(v2)
```

Usefull attributes on form

```python
# number of dimensions
print(d2.ndim)

# number of cells
print(d2.size)
```

```python
# size is the product of dimensions
len(d2.shape) == d2.ndim
```

```python
#double check size with mul
from operator import mul
from functools import reduce
d2.size == reduce(mul, d2.shape, 1)
```

We can then expand nupy arrays to higher dimensions

```python
shape = (2, 3, 4)
size = reduce(mul, shape)
print(size)
# you can use arange
data = np.arange(size)
print(data)
```

```python
d3 = data.reshape(shape)
d3
```

So with higher dimensions

```python
# same thing with bigger dim
shape = (2, 3, 4, 5)
size = reduce(mul, shape) # product of the 4 numbers in shape
size
```

```python
data = np.arange(size)
print(data)

# visualise the array d4
# from building blocks of base
# 4 rows and et 5 columns

d4 = data.reshape(shape)
d4
```

We can always add some 1 in the shape, it won't change the product of dimensions

```python
#try
d2.shape
d2.reshape(2, 1, 3)
d2.reshape(2, 3, 1)
d2.reshape((1, 2, 3))
d2.reshape((1, 1, 1, 1, 2, 3))
```

To make computation, we sometimes need to flatten multidimensional arrays. Two methods:
    - ravel
    - flatten

```python
rav_d2 = d2.ravel()
rav_d2
```

```python
flat_d2 = d2.flatten()
flat_d2
```
